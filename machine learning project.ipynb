### Impute median for missing value

```python
X_train = dtrain.iloc[:,0:148]
Y_train = dtrain.iloc[:,148]
X_valid = dvalid.iloc[:,0:148]
Y_valid = dvalid.iloc[:,148]
X_test=dtest
```

```python
X_test['L6'].fillna((X_train['L6'].median()), inplace=True)
X_valid['L6'].fillna((X_train['L6'].median()), inplace=True)
X_train['L6'].fillna((X_train['L6'].median()), inplace=True)

X_test['M6'].fillna((X_train['M6'].median()), inplace=True)
X_valid['M6'].fillna((X_train['M6'].median()), inplace=True)
X_train['M6'].fillna((X_train['M6'].median()), inplace=True)

X_test['N6'].fillna((X_train['N6'].median()), inplace=True)
X_valid['N6'].fillna((X_train['N6'].median()), inplace=True)
X_train['N6'].fillna((X_train['N6'].median()), inplace=True)

X_test['T6'].fillna((X_train['T6'].median()), inplace=True)
X_valid['T6'].fillna((X_train['T6'].median()), inplace=True)
X_train['T6'].fillna((X_train['T6'].median()), inplace=True)

X_test['U6'].fillna((X_train['U6'].median()), inplace=True)
X_valid['U6'].fillna((X_train['U6'].median()), inplace=True)
X_train['U6'].fillna((X_train['U6'].median()), inplace=True)

X_test['V6'].fillna((X_train['V6'].median()), inplace=True)
X_valid['V6'].fillna((X_train['V6'].median()), inplace=True)
X_train['V6'].fillna((X_train['V6'].median()), inplace=True)

X_test['W6'].fillna((X_train['W6'].median()), inplace=True)
X_valid['W6'].fillna((X_train['W6'].median()), inplace=True)
X_train['W6'].fillna((X_train['W6'].median()), inplace=True)

X_test['X6'].fillna((X_train['X6'].median()), inplace=True)
X_valid['X6'].fillna((X_train['X6'].median()), inplace=True)
X_train['X6'].fillna((X_train['X6'].median()), inplace=True)
```

### Optional Step: subset observation

```python
chosen_idx = np.random.choice(len(X_train), replace = True, size = 50000) 
X_train = X_train.iloc[chosen_idx]
Y_train = Y_train.iloc[chosen_idx]
```

### Optional Step: standardize

```python
std_scale = preprocessing.StandardScaler().fit(X_train)
X_train = std_scale.transform(X_train)
X_valid  = std_scale.transform(X_valid)
X_test  = std_scale.transform(X_test)
```
